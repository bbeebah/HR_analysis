```
rename_col = {
    'Exception Description (0033)':'HireType',
    'Statistics Indicator Description (0033)':'Indicator'
}
HireType.rename(rename_col, axis=1, inplace=True)

HireType = HireType[HireType['Indicator'] == 'Hire type']
```
